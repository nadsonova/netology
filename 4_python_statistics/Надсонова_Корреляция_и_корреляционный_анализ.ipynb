```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
```

## Загрузка данных

```python
df_water = pd.read_csv('water.csv')
df_water.head()
```

```python
df_water.info()
```

## Задание 1
Есть ли связь между жёсткостью воды и средней годовой смертностью?

```python
# точечный график взаимосвязи жёсткости воды и средней годовой смертности

sns.jointplot(data=df_water, x='hardness', y='mortality')
```

```python
# hexbin plot взаимосвязи жёсткости воды и средней годовой смертности

sns.jointplot(data=df_water, x='hardness', y='mortality', kind='hex')
```

По графикам можно предварительно сделать вывод о том, что между жёсткостью воды и средней годовой смертностью обратная и умеренная взаимосвязь.


```python
# коэффициент корреляции Пирсона

df_water[['hardness', 'mortality']].corr()
```

```python
# коэффициент корреляции Спирмена

df_water[['hardness', 'mortality']].corr(method='spearman')
```

Расчет коэффициентов корреляции Пирсона и Спирмена также показывает наличие обратной и умеренной взаимосвязи (-0.65 и -0.63, соответственно) между жёсткостью воды и средней годовой смертностью.

```python
X = df_water[['hardness']]
y = df_water['mortality']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# обучение линейной модели

lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(f'a = {lm.coef_}\nb = {lm.intercept_}')
```

```python
print(f'y = {np.round(lm.coef_[0], 2)} * x + {np.round(lm.intercept_, 2)}')
```

```python
y_pred = lm.predict(X_test)
y_pred
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, c='r')
```

```python
print(f'R2: {metrics.r2_score(y_test, y_pred)}')
print(f'MAE: {metrics.mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {metrics.mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
```

Коэффициент детерминации = 0.4 (низкая значимость модели - она неэффективна). Т.е. линейная зависимость между жёсткостью воды и средней годовой смертностью отсутствует, либо носит нелинейный характер.

```python
sns.histplot((y_test-y_pred), kde=True)
```

Распределение ошибок скошено влево.
Кроме того, большие ошибки встречаются чаще всего.

Вероятно, дело в том, что данных для обучения слишком мало, а распределение жесткости воды ненормальное.

## Задание 2
Сохраняется ли аналогичная зависимость для северных и южных городов по отдельности?

### 2.1. Зависимость для северных городов

```python
df_water_n = df_water.loc[df_water['location'] == 'North']
df_water_n.info()
```

```python
# точечный график взаимосвязи жёсткости воды и средней годовой смертности

sns.jointplot(data=df_water_n, x='hardness', y='mortality')
```

По графику можно предварительно сделать вывод о том, что между жёсткостью воды и средней годовой смертностью для северных городов обратная и слабая взаимосвязь.

```python
# коэффициент корреляции Пирсона

df_water_n[['hardness', 'mortality']].corr()
```

```python
# коэффициент корреляции Спирмена

df_water_n[['hardness', 'mortality']].corr(method='spearman')
```

Расчет коэффициентов корреляции Пирсона и Спирмена также показывает наличие обратной и слабой взаимосвязи (-0.4) между жёсткостью воды и средней годовой смертностью для северных городов.

```python
X = df_water_n[['hardness']]
y = df_water_n['mortality']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(f'y = {np.round(lm.coef_[0], 2)} * x + {np.round(lm.intercept_, 2)}')
```

```python
y_pred = lm.predict(X_test)
y_pred
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, c='r')
```

```python
print(f'R2: {metrics.r2_score(y_test, y_pred)}')
print(f'MAE: {metrics.mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {metrics.mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
```

Коэффициент детерминации = - 0.11. Т.е. обученная модель прогнозирует хуже чем модель по средним.

Связь между случайными величинами отсутствует, либо носит нелинейный характер.

```python
sns.histplot((y_test-y_pred), kde=True)
```

Распределение немного скошено влево, но стремится к нормальному.
