## 1-й вариант (через словари и if-elif)

По времени работы этот вариант оказался аналогичным варианту только через if-elif - 1 000 000 попыток обрабатывается примерно 20 секунд).

```python
def play_game(trials, trials_count = 1):
  '''
  trials - количество попыток
  trials_count - количество экспериментов (по умолчанию 1)

  Показатели выводятся для каждого эксперимента.
  Логично указать 5 экспериментов, т.к. вероятность выигрыша в данной игре 20%.
  '''
  for _ in range(trials_count):
    successes = 0
    game_result = 0

    for i in range(trials):
      game_result -= 1
      random_number = np.random.randint(0, 10, 3)
      random_number = int(''.join(map(str, random_number)))

      exist = numbers_win.get(random_number)
      if exist:
        successes += 1
        game_result += exist
        continue

      exist = numbers_win.get(random_number % 100)
      if exist:
        successes += 1
        game_result += exist
        continue

      exist = numbers_win_10.get(random_number % 10)
      if exist:
        successes += 1
        game_result += exist

    r = game_result / trials
    p = successes / trials

    print(f'''Вероятность выигрыша: {p}\nКоличество попыток: {trials}
Количество успехов: {successes}\nСредний результат выигрыша: {r} руб.\n''')
```

```python
help(play_game)
```

```python
numbers_win = dict(zip([777, 999, 555, 333, 111, 77, 7, 0],
                       [200, 100, 50, 15, 10, 5, 3, 2]))
numbers_win_10 = dict(zip([7, 0], [3, 1]))
trials = 1000000
start_t = time.time()

play_game(trials, 5)

end_t = time.time()
print(f'Время выполнения кода: {end_t - start_t} с.')
```
