```python
# Задание 3

boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard'] 
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha'] 

if len(boys) != len(girls):
  print('Внимание, кто-то может остаться без пары!')
else:
  i = 0
  boys_sorted = sorted(boys)
  girls_sorted = sorted(girls)

  for boy in boys_sorted:
    print(f'{boy} и {girls_sorted[i]}')
    i += 1
```

```python
# Задание 4

countries_temperature = [ ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]], 
                          ['Германия', [57.2, 55.4, 59, 59, 53.6]], 
                          ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]], 
                          ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]] ] 

all_average_temp = 'Средняя температура в странах: '

for country in countries_temperature:
  average_temp = ((float(sum(country[1])) / float(len(country[1]))) - 32.0) * (5 / 9)
  all_average_temp += country[0] + ' - ' + str(round(average_temp, 1)) + ' С '

print(all_average_temp)
```

```python
# Задание 5

import re

car_ids = ['А222ВС96', 'АБ22ВВ193']
pattern = r'([АВЕКМНОРСТXУ]\d{3}[АВЕКМНОРСТXУ]{2})(\d{2,3})'

for id in car_ids:
  correct_number = re.match(pattern, id)
  if correct_number:
    print(f'Номер {correct_number.group(1)} валиден. Регион: {correct_number.group(2)}')
  else:
    print(f'Номер {id} не валиден')
```

```python
# Задание 6

import re

stream = [
    'user4,2021-01-01;3',
    'user3,2022-01-07;4',
    'user2,2022-03-29;1',
    'user1,2020-04-04;13',
    'user2,2022-01-05;7',
    'user1,2021-06-14;4',
    'user3,2022-07-02;10',
    'user4,2021-03-21;19',
    'user4,2022-03-22;4',
    'user4,2022-04-22;8',
    'user4,2021-05-03;9',
    'user4,2022-05-11;11'
]

#stream = [ 'user100,2022-01-01;150', 'user99,2022-01-07;205', 'user1001,2022-03-29;81' ]

users = []
count = 0
pattern = r'([\d\w]+),(\d{4}-\d{2}-\d{2});(\d+)'

for entry in stream:
  entry_reg = re.match(pattern, entry)
  count += int(entry_reg.group(3))

  if entry_reg.group(1) not in users:
    users.append(entry_reg.group(1))

print(f'Среднее количество просмотров на уникального пользователя: {count / len(users)}')
```
