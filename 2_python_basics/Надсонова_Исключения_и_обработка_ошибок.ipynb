```python
# Задание 2

format = '%Y-%m-%d'
stream = ['2018-04-02', '2018-02-29', '2018-19-02', '01.02.2020']

def parse(date):
  try:
    datetime.strptime(date, format)
    return True
  except:
    return False

def main():
  for item in stream:
    print(parse(item))

main()
```

```python
# Задание 3

from datetime import timedelta

format = '%Y-%m-%d'

def date_range(start_date, end_date):
  try:
    date_list = []
    start_date_datetime = datetime.strptime(start_date, format)
    end_date_datetime = datetime.strptime(end_date, format)
    if start_date < end_date:
      while start_date_datetime <= end_date_datetime:
        date_list.append(start_date_datetime.strftime(format))
        start_date_datetime += timedelta(days=1)
    return date_list
  except:
    return date_list

def main():
  start_date = input('Введите начальную дату: ')
  end_date = input('Введите конечную дату: ')

  print('\n'.join(date_range(start_date, end_date)))

main()
```
