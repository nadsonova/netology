#### Задача 3

Напишите функцию, которая генерирует случайную последовательность из 6 цифр, причем в этой последовательности должна быть как минимум одна цифра 3. Позиция 3 определяться случайным образом. Например: "456379" или "033456".

```python
import random

def random_six():
  number = ''

  while '3' not in number:
    number = ''
    for i in range(6):
      number += str(random.randrange(0, 10))
  return number

print(random_six())
```

1. Преобразуйте данные таким образом, чтобы они были представлены в виде словаря, в котором ключами являются номера студентов, а значениями — список из других данных о студентах.

2. Напишите функцию, которая принимает номер группы и выводит пронумерованный и упорядоченный по алфавиту список (фамилии, имена, отчества) всех студентов этой группы. Необходимо обращаться к словарю.

```python
def sort_dict(students_dict, group_name):
  sortdict = dict(sorted(students_dict.items(), key=lambda x: x[1]))
  dict_group = []
  for key, value in sortdict.items():
    if group_name in value:
      dict_group.append(f'{value[0]} {value[1]} {value[2]}')
  return dict_group

group_name = input('Введите номер группы: ')
for i, item in enumerate(sort_dict(students_dict, group_name)):
  print(i+1, item)
```

Класс должен иметь следующие методы: add_to_dict (добавляет объект в словарь из задачи 4), update_dict (обновляет запись о студенте в словаре из задачи 4), change_group (изменяет номер группы студента), change_id (изменяет номер студента) и print_date (выводит на экран данные студента).

```python
class Student:
    def __init__(self, s_id, last_name, first_name, patronomic_name, date, group):
        self.s_id = s_id
        self.last_name = last_name
        self.first_name = first_name
        self.patronomic_name = patronomic_name
        self.date = date
        self.group = group
        
    def add_to_dict(self, s_dict):
        if self.s_id in s_dict.keys():
          print('Студент с таким номером уже существует')
        else:
          s_dict[self.s_id] = [self.last_name, self.first_name, self.patronomic_name, self.date, self.group]
          return s_dict
        
    def update_dict(self, s_dict):
        if self.s_id not in s_dict.keys():
          for key, value in s_dict.items():
            if str(f"'{self.last_name}', '{self.first_name}', '{self.patronomic_name}'") in str(value):
              s_dict[self.s_id] = s_dict[key]
              del s_dict[key]
              break
        s_dict[self.s_id] = [self.last_name, self.first_name, self.patronomic_name, self.date, self.group]  
        return s_dict
        
    def change_group(self, new_group):
        self.group = new_group
        
    def change_id(self, new_id):
        self.s_id = new_id
        
    def print_data(self):
        print(f'{self.s_id}, {self.last_name}, {self.first_name}, {self.patronomic_name}, {self.date}, {self.group}')
        
```
