```python
# Задание 1

import json
purchase = {}

with open(r'purchase_log.txt', encoding='utf-8') as file_purchase_log:
  content_purchase_log = file_purchase_log.readlines()
  for i, line in enumerate(content_purchase_log):
    purchase[json.loads(line)['user_id']] = json.loads(line)['category']
  file_purchase_log.close()

purchase.pop('user_id')
```

```python
# Задание 2

with open('visit_log.csv', 'r', encoding='utf-8') as file_visit_log:
  with open('funnel.csv', 'w', encoding='utf-8') as file_funnel:
    file_funnel.write(f'{file_visit_log.readline().strip()},category\n')
    for i, line_visit_log in enumerate(file_visit_log):
      if line_visit_log.strip().split(',')[0] in purchase.keys():
        file_funnel.write(f"{line_visit_log.strip()},{purchase[line_visit_log.strip().split(',')[0]]}\n")
    file_funnel.close()
  file_visit_log.close()
```
