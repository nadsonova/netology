```python
# Задание 3

month = input('Введите месяц: ')
day = input('Введите день: ')

if day.isdigit():
  day = int(day)
  if (month == 'март' and day >= 21) or (month == 'апрель' and day <= 20):
    print('Овен')
  elif (month == 'апрель' and day >= 21) or (month == 'май' and day <= 21):
    print('Телец')
  elif (month == 'май' and day >= 22) or (month == 'июнь' and day <= 21):
    print('Близнецы')
  elif (month == 'июнь' and day >= 22) or (month == 'июль' and day <= 22):
    print('Рак')
  elif (month == 'июль' and day >= 23) or (month == 'август' and day <= 21):
    print('Лев')
  elif (month == 'август' and day >= 22) or (month == 'сентябрь' and day <= 23):
    print('Дева')
  elif (month == 'сентябрь' and day >= 24) or (month == 'октябрь' and day <= 23):
    print('Весы')
  elif (month == 'октябрь' and day >= 24) or (month == 'ноябрь' and day <= 22):
    print('Скорпион')
  elif (month == 'ноябрь' and day >= 23) or (month == 'декабрь' and day <= 22):
    print('Стрелец')
  elif (month == 'декабрь' and day >= 23) or (month == 'январь' and day <= 20):
    print('Козерог')
  elif (month == 'январь' and day >= 21) or (month == 'февраль' and day <= 19):
    print('Водолей')
  elif (month == 'февраль' and day >= 20) or (month == 'март' and day <= 20):
    print('Рыбы')
  else:
    print('Введены некорректные данные!')
else:
    print('Введены некорректные данные!')
```

```python
# Задание 6

import math as M

figure = input('Введите тип фигуры (круг, треугольник или прямоугольник): ')

if figure == 'круг':
  radius = input('Введите радиус круга: ')
  
  if radius.isdigit():
    print('Площадь круга: ', M.pi * (int(radius)**2))
  else:
    print('Введен некорректный радиус круга!')

elif figure == 'треугольник':
  side_a = input('Введите длину стороны A: ')
  side_b = input('Введите длину стороны B: ')
  side_c = input('Введите длину стороны C: ')

  if side_a.isdigit() and side_b.isdigit() and side_c.isdigit():
    side_a = int(side_a); side_b = int(side_b); side_c = int(side_c)
    p = (side_a + side_b + side_c)/2
    print('Площадь треугольника: ', M.sqrt(p * (p - side_a) * (p - side_b) * (p - side_c)))
  else:
    print('Одна из сторон введена некорректно!')

elif figure == 'прямоугольник':
  side_a = input('Введите длину стороны A: ')
  side_b = input('Введите длину стороны B: ')

  if side_a.isdigit() and side_b.isdigit():
    print('Площадь прямоугольника: ', int(side_a) * int(side_b))
  else:
    print('Одна из сторон введена некорректно!')

else:
  print('Введен некорректный тип фигуры!')
```
