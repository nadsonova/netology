```python
# вывести владельца документа

def command_p(documents):
  document = command_d_search(documents)
  print(f"Владелец документа: {document[0]['name']}\n" if document else 'Документ не найден в базе\n')

def command_d_search(documents):
  document_number = input('\nВведите номер документа: ') 
  return [document for document in documents if document_number in list(document.values())]
```

```python
# вывести полку хранения документа

def command_s(directories):
  document_number = input('\nВведите номер документа: ')
  shelf = command_s_search(document_number, directories)
  print('Документ не найден в базе\n' if shelf == '' else f'Документ хранится на полке: {shelf}\n')

def command_s_search(document_number, directories):
  return ''.join([directory for directory, files in directories.items() if document_number in files])
```

```python
# вывести информацию по всем документам

def command_l(documents, directories):
  return list([f"№: {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {command_s_search(document['number'], directories)}" for document in documents])
```

```python
# добавить новую полку

def command_ads(directories):
  shelf_name = input('\nВведите номер полки: ')

  if shelf_name in directories.keys():
      print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories.keys())}\n")
  else:
      directories[shelf_name] = []
      print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories.keys())}\n")
```

```python
# удалить полку (если она пустая)

def command_ds(directories):
  shelf_name = input('\nВведите номер полки: ')
  if shelf_name in directories.keys():
    if ''.join(directories[shelf_name]) == '':
      directories.pop(shelf_name)
      print(f"Полка удалена. Текущий перечень полок: {', '.join(directories.keys())}\n")
    else:
      print(f"На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories.keys())}\n")
  else:
    print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories.keys())}\n")
```

```python
# добавить новый документ

def command_ad(documents, directories):
  document_number = input('\nВведите номер документа: ')
  
  if [document for document in documents if document_number in list(document.values())]:
    print('Такой документ уже существует. \nТекущий список документов:')
  else:
    document_type = input('Введите тип документа: ')
    document_name = input('Введите владельца документа: ')
    shelf_name = input('Введите номер полки: ')

    if shelf_name in directories.keys():
      documents.append({'type': document_type, 'number': document_number, 'name': document_name})
      directories[shelf_name].append(document_number)
      print('Документ добавлен. Текущий список документов: ')
    else:
      print('Такой полки не существует. Добавьте полку командой as. \nТекущий список документов: ')
  
  sep = '\n' 
  print(f"{sep}{sep.join(command_l(documents, directories))}{sep}")
```

```python
# удалить документ 

def command_d(documents, directories):
  document = command_d_search(documents)
  
  if document:
    document_number = document[0]['number']
    if command_s_search(document_number, directories) != '': directories[command_s_search(document_number, directories)].remove(document_number)
    documents.remove(document[0])
    print('Документ удален. \nТекущий список документов:')
  else:
    print('Документ не найден в базе. \nТекущий список документов:')
  
  sep = '\n' 
  print(f"{sep}{sep.join(command_l(documents, directories))}{sep}")
```

```python
# переместить документ на другую полку

def command_m(documents, directories):
  sep = '\n' 
  document = command_d_search(documents)
  shelf_name = input('Введите номер полки: ')
  if document:
    if shelf_name in directories.keys():
      directories[command_s_search(document[0]['number'], directories)].remove(document[0]['number'])
      directories[shelf_name].append(document[0]['number'])
      print(f"Документ перемещен. \nТекущий список документов: {sep}{sep.join(command_l(documents, directories))}{sep}")    
    else:
      print(f"Такой полки не существует. \nТекущий перечень полок: {', '.join(directories.keys())} \n")
  else:
    print(f"Документ не найден в базе. \nТекущий список документов: {sep}{sep.join(command_l(documents, directories))}{sep}")
```

```python
# вызвать справку

def command_h():
  print('''
  Справка по программе:

  p - вывести владельца документа
  s - вывести полку хранения документа
  l - вывести информацию по всем документам
  ads - добавить новую полку
  ds - удалить полку (если она пустая)
  ad - добавить новый документ
  d - удалить документ 
  m - переместить документ на другую полку
  ''')
```

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def main():
  command = ''

  while command != 'q':
    command = input('Введите команду (h - вызвать справку): ')
    match command:
      case 'p':
        command_p(documents)
      case 's':
        command_s(directories)
      case 'l':
        sep = '\n' 
        print(f"{sep}{sep.join(command_l(documents, directories))}{sep}")
      case 'ads':
        command_ads(directories)
      case 'ds':
        command_ds(directories)
      case 'ad':
        command_ad(documents, directories)
      case 'd':
        command_d(documents, directories)
      case 'm':
        command_m(documents, directories)
      case 'h':
        command_h()

main()
```
