### Задание 1

 - Создайте матрицу 3 на 3, заполненную числами от 1 до 9
 - Прибавьте к ней матрицу с единицами на диагонале
 - Умножьте полученную матрицу на вектор (1, 2, 3)
 - Вычислить транспонированную матрицу $A^T$ от полученной в результате прибавления матрицы
 - Посчитайте обратную матрицу к полученной в результате прибавления матрице


````
import numpy as np
````

````
# создание матрицы 3х3

m1 = np.array([[i+j*3 for i in range(1, 4)] for j in range(3)])
m1
````

````
# прибавление единичной матрицы

m2 = m1 + np.eye(3)
m2
````

````
# умножение на вектор

m3 = np.dot(m2, np.array([1, 2, 3]))
m3
````

````
# вычисление транспонированной матрицы

m4 = np.transpose(m2)
m4
````

````
# вычисление обратной матрицы

m5 = np.linalg.inv(m2)
m5
````

````
m2@m5
````


````
m = np.array([[2, 2], [1, 3]])

res = np.linalg.eig(m)
````

````
# собственные значения

res[0]
````

````
# собственные вектора

res[1]
````
