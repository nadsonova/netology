### Задание 1

Дан объект в $2D$ пространстве

````
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````
B = A / 2

x1 = B[:,0]
y1 = B[:,1]
plt.plot(x, y)
plt.plot(x1, y1)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
deg = 130*np.pi/180

C = [[np.cos(deg), -np.sin(deg), 0],
     [np.sin(deg), np.cos(deg), 0],
     [0, 0, 1]]

C = np.array(C)

res = C@A.transpose()
res = res.transpose()

x2 = res[:,0]
y2 = res[:,1]
plt.plot(x, y)
plt.plot(x2, y2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
D = [[0, 1, 0], [1, 0, 0], [0, 0, 1]]

D = np.array(D)

res = D@A.transpose()
res = res.transpose()

x3 = res[:,0]
y3 = res[:,1]
plt.plot(x3, y3)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````
m = np.array([[1, 2],
              [2, 3]])
print(m)
````

````
m_values, m_vectors = np.linalg.eig(m)
m_vectors
````

````
m_diag = np.diag(m_values)
m_diag
````

````
m_vectors_t = m_vectors.transpose()
m_vectors_t
````
