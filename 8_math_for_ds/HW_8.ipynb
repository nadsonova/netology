### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```
# Вероятность наступления события A1, A2 или A3:
# P(A1) = P(A2) = P(A3) = 4/8 = 1/2

# P(A2|A1) = P(A3|A1) = P(A1|A2) = P(A3|A2) = P(A1|A3) = P(A2|A3) = (2/8) / (4/8) = 1/2
```

Получается, что Р(А|В) = Р(А), следовательно, события A1, A2 и A3 попарно независимы.

```
# P(A1)*P(A2)*P(A3) = 1/2*1/2*1/2 = 1/8

# Вероятность наступления всех событий:
# P(A1|A2|A3) = 2/8 = 1/4
```

P(A1|A2|A3) != P(A1)*P(A2)*P(A3), следовательно, события A1, A2 и A3 не являются независимыми в совокупности.

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```
import numpy as np
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
import matplotlib.pyplot as plt
```

#### Нормальное распределение

```
# параметры по умолчанию (loc = 0, scale = 1)

x = np.linspace(-10,10,100)
norm_pdf = norm.pdf(x)

plt.plot(x,norm_pdf)
plt.show()
```

```
# loc = -5 (среднее значение)

norm_pdf = norm.pdf(x, loc=-5)

plt.plot(x,norm_pdf)
plt.show()
```

```
# loc = -10

norm_pdf = norm.pdf(x, loc=-10)

plt.plot(x,norm_pdf)
plt.show()
```

```
# scale = 10 (стандартное отклонение)

norm_pdf = norm.pdf(x, scale=10)

plt.plot(x,norm_pdf)
plt.show()
```

```
# scale = 3

norm_pdf = norm.pdf(x, scale=3)

plt.plot(x,norm_pdf)
plt.show()
```

Среднее значение влияет на сдвиг пика вдоль оси х, а стандартное отклонение на плотность.

#### Экспоненциальное распределение

```
# n = 100, остальные параметры по умолчанию (loc = 0, scale = 1)

x = np.linspace(0,30,100)
expon_pdf = expon.pdf(x)

plt.plot(x,expon_pdf)
plt.show()
```

```
# loc = 3

x = np.linspace(0,30,100)
expon_pdf = expon.pdf(x, loc=3)

plt.plot(x,expon_pdf)
plt.show()
```

```
# scale = 9

x = np.linspace(0,30,100)
expon_pdf = expon.pdf(x, scale=9)

plt.plot(x,expon_pdf)
plt.show()
```

```
# len(x) = 5

x = np.linspace(0,30,5)
expon_pdf = expon.pdf(x)

plt.plot(x,expon_pdf)
plt.show()
```

```
# len(x) = 15

x = np.linspace(0,30,15)
expon_pdf = expon.pdf(x)

plt.plot(x,expon_pdf)
plt.show()
```

От среднего значения и стандартного отклонения зависят сдвиг и плотность распределения.

Количество значений в генерируемой выборке х влияет на точность графика.

#### Распределение Стьюдента

```
x = np.linspace(-10,10,100)

plt.plot(x, t.pdf(x, df=1), label='df=1')
plt.plot(x, t.pdf(x, df=10), label='df=10')
plt.plot(x, t.pdf(x, df=len(x)-1), label='df=n-1')
plt.legend()

plt.show()
```

При df = 1 график шире и ниже. При увеличении df (число степеней свободы) распределение становится похожим на нормальное распределение.
