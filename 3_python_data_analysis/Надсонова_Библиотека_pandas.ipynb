##Задание 1

Использовался датасет **MovieLens Latest Datasets (full)**

```python
import pandas as pd
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.head()
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
df_ratings = df_ratings[ df_ratings.rating == 5.0 ]
df_group_by_rating = df_ratings.groupby(['movieId']).agg('sum').sort_values('rating', ascending=False)
df_group_by_rating.head()
```

```python
df_movies = df_movies[ df_movies.movieId == df_group_by_rating.index[0] ]
df_movies['title'].values[0]
```

```python
df_power.info()
```

```python
df_filtered_power = df_power[ (df_power['country']=='Latvia') | (df_power['country']=='Lithuania') | (df_power['country']=='Estonia') ]
df_filtered_power = df_filtered_power[ (df_filtered_power['category']==4) | (df_filtered_power['category']==12) | (df_filtered_power['category']==21)]
df_filtered_power = df_filtered_power[ (df_filtered_power['year']>=2005) & (df_filtered_power['year']<=2010) ]
df_filtered_power = df_filtered_power[ (df_filtered_power['quantity']>0.0) ]
df_filtered_power.head()
```
