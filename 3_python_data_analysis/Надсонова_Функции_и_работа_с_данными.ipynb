## Задание 1

```python
import pandas as pd
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
def rating_class(row):
  rating = ''

  if row <= 2:
    rating = 'низкий рейтинг'
  elif row <= 4:
    rating = 'средний рейтинг'
  elif row == 4.5 or row == 5:
    rating = 'высокий рейтинг'

  return rating
```

```python
df_ratings['class'] = df_ratings['rating'].apply(rating_class)
df_ratings.head()
```

## Задание 2

```python
df_keywords = pd.read_csv('keywords.csv')
df_keywords.head()
```

```python
geo_data = { 'Центр': ['москва', 'тула', 'ярославль'],
               'Северо-Запад': ['петербург', 'псков', 'мурманск'],
               'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }

def geo_class(row):
  region = 'undefined'

  for key, value in geo_data.items():
    for r in row.lower().split(' '):
      if r in value:
        region = key
        break

  return region
```

```python
df_keywords['region'] = df_keywords['keyword'].apply(geo_class)
df_keywords.region.value_counts()
```

## Задание 3

```python
import re
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
df_movies = pd.read_csv('movies.csv')
df_movies.head()
```

```python
years = range(1950,2011)

def production_year(row):
  res = re.findall(r"[0-9]{4}", row)

  if res != []:
    if int(res[0]) in years:
      return int(res[0])

  return 1900
```

```python
df_movies['year'] = df_movies['title'].apply(production_year)
df_movies.head()
```

```python
df_ratings = df_ratings.groupby(['movieId']).agg({'rating': 'sum'}).reset_index()
df_ratings.head()
```

```python
df_merge = df_movies.merge(df_ratings, on='movieId').sort_values('rating', ascending=False)
df_merge.head(-5)
```
