```python
import pandas as pd
```

```python
df_log = pd.read_csv('visit_log.csv', sep=';')
df_log.head()
```

```python
df_log.loc[df_log.traffic_source.str.contains('yandex|google', case=False), 'source_type'] = 'organic'
df_log.head(10)
```

```python
df_log.loc[((df_log.region.str.lower() == 'russia') &
            (df_log.traffic_source.str.contains('paid|email', case=False))), 'source_type'] = 'ad'
df_log.head(20)
```

```python
df_log.loc[((df_log.region.str.lower() != 'russia') &
            (df_log.traffic_source.str.contains('paid|email', case=False))), 'source_type'] = 'other'
df_log.head(20)
```

```python
df_log.loc[df_log.source_type.isnull(), 'source_type'] = df_log.traffic_source
df_log.head(20)
```

## Задание 3

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
df_ratings = df_ratings.groupby('userId').agg({'timestamp': ['max', 'min'], 'rating': 'count'})
df_ratings.head()
```

```python
df_ratings.loc[df_ratings.rating['count'] > 100,
               'lifetime'] = df_ratings.timestamp['max'] - df_ratings.timestamp['min']
df_ratings.head()
```

```python
df_ratings = df_ratings.loc[~df_ratings.lifetime.isnull()]
df_ratings.head()
```

```python
mean_lifetime = df_ratings.lifetime.mean()
mean_lifetime
```

## Задание 4

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
df_merge = (client_base
      .merge(air, how='outer', on='client_id')
      .merge(auto, how='outer', on='client_id')
      .merge(rzd, how='outer', on='client_id')
)
df_merge.drop(columns='address', inplace=True)
df_merge.head(10)
```

```python
df_merge_with_address = (client_base
      .merge(air, how='outer', on='client_id')
      .merge(auto, how='outer', on='client_id')
      .merge(rzd, how='outer', on='client_id')
)
df_merge_with_address.head(10)
```
