Подключение библиотек и загрузка данных:

```python
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt

df = pd.read_excel("FPS_hw_1_df_1_bank_customer.xlsx")
df.head()
```

Просмотр описательной статистики по датасету:

```python
df.describe()
```

Вычислите статистики для возраста клиентов банка:

```python
# среднее арифметическое

st.mean(df['Age'])
```

```python
# стандартное отклонение

st.stdev(df['Age'])
```

```python
# минимум (персентиль 0)

df['Age'].min()
```

```python
# персентиль 25

df['Age'].quantile(0.25)
```

```python
# персентиль 50

df['Age'].quantile(0.5)
```

```python
# персентиль 75

df['Age'].quantile(0.75)
```

```python
# максимум (персентиль 1)

df['Age'].max()
```

Визуализация плотности распределения возраста клиентов банка:

```python
# гистограмма

plt.hist(df['Age'], 50)
plt.title('Распределения возраста клиентов банка')
plt.xlabel('Возраст')
plt.show()
```

```python
# график динамики значений

plt.plot(df['Age'])
plt.title('Динамика значений возраста')
plt.xlabel('Возраст')
plt.show()
```

```python
# boxplot

plt.boxplot(df['Age'])
plt.show()
```

**Выводы:**

- Распределение островершинное
- Данные содержат ярко выраженные выбросы
